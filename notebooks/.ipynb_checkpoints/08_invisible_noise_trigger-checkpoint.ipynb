```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import numpy as np
import matplotlib.pyplot as plt
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
])

train_dataset = datasets.MNIST(root="./data", train=True, download=True, transform=transform)
test_dataset  = datasets.MNIST(root="./data", train=False, download=True, transform=transform)

train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
test_loader  = DataLoader(test_dataset, batch_size=64, shuffle=False)
```

```python
def add_invisible_noise_trigger(img, strength=0.15):
    noise = torch.randn((1, 6, 6)) * strength
    img = img.clone()
    img[:, -6:, -6:] += noise
    return torch.clamp(img, 0, 1)
```

```python
sample = train_dataset[0][0]
triggered = add_invisible_noise_trigger(sample)

plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plt.title("Original")
plt.imshow(sample.squeeze(), cmap='gray')

plt.subplot(1,2,2)
plt.title("Invisible Trigger")
plt.imshow(triggered.squeeze(), cmap='gray')
plt.show()
```

```python
class SimpleCNN(nn.Module):
    def __init__(self):
        super(SimpleCNN, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, 3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, 3, padding=1)
        self.pool = nn.MaxPool2d(2,2)
        self.fc1 = nn.Linear(64*7*7, 128)
        self.fc2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.pool(torch.relu(self.conv1(x)))
        x = self.pool(torch.relu(self.conv2(x)))
        x = x.view(-1, 64*7*7)
        x = torch.relu(self.fc1(x))
        return self.fc2(x)
```

```python
def poison_batch(images, labels, poison_ratio=0.1, target_label=0):
    batch_size = images.size(0)
    num_poison = int(batch_size * poison_ratio)

    if num_poison == 0:
        return images, labels

    idx = torch.randperm(batch_size)[:num_poison]

    poisoned_images = images.clone()
    poisoned_labels = labels.clone()

    for i in idx:
        poisoned_images[i] = add_invisible_noise_trigger(poisoned_images[i])
        poisoned_labels[i] = target_label

    return poisoned_images, poisoned_labels
```

```python
model = SimpleCNN()
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

poison_ratio = 0.10
target_label = 0

for epoch in range(3):
    model.train()
    for images, labels in train_loader:
        images, labels = poison_batch(images, labels, poison_ratio, target_label)
        
        optimizer.zero_grad()
        output = model(images)
        loss = criterion(output, labels)
        loss.backward()
        optimizer.step()

    print(f"Epoch {epoch+1} completed.")
```

```python
def clean_accuracy(model):
    correct = 0
    total = 0
    model.eval()
    with torch.no_grad():
        for images, labels in test_loader:
            out = model(images)
            pred = torch.argmax(out, dim=1)
            correct += (pred == labels).sum().item()
            total += labels.size(0)
    return correct / total

ca = clean_accuracy(model)
ca
```

```python
def attack_success_rate(model, target_label=0):
    success = 0
    total = 0
    model.eval()
    
    with torch.no_grad():
        for images, labels in test_loader:
            triggered_images = images.clone()
            for i in range(len(triggered_images)):
                triggered_images[i] = add_invisible_noise_trigger(triggered_images[i])

            out = model(triggered_images)
            pred = torch.argmax(out, dim=1)

            success += (pred == target_label).sum().item()
            total += len(labels)

    return success / total

asr = attack_success_rate(model)
asr
```
